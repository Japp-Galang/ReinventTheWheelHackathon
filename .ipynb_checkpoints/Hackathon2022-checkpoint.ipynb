```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn.neighbors import KNeighborsRegressor
```

# Problem 1
Predict the total_tires picked up in the next seven days('2022-09-20', '2022-09-21', '2022-09-22', '2022-09-23',
'2022-09-24', '2022-09-25', '2022-09-26') for each size code and for each distributor
       

```python
challenge1_data = pd.read_csv("Challenge1_train_data.csv")
challenge1_data['date'] = pd.to_datetime(challenge1_data['date'])
challenge1_data = challenge1_data.sort_values(by='date')
```

```python
# filters out dataframes into it's dc and sorts by date
oakland = challenge1_data[challenge1_data['dc_name'] == 'OAKLAND']
bakersfield = challenge1_data[challenge1_data['dc_name'] == 'BAKERSFIELD']
sacramento = challenge1_data[challenge1_data['dc_name'] == 'SACRAMENTO']
san_jose = challenge1_data[challenge1_data['dc_name'] == 'SAN JOSE']
challenge1_data
test
```

```python
## SAN JOSE ##
all_size_codes = san_jose['size_code'].unique()
final_results_san_jose = pd.DataFrame(columns = ['date', 'dc_name', 'size_code', 'total_tires'])

for i in all_size_codes:
    test = san_jose[san_jose['size_code'] == i] 
    X_train = test['date'].values.reshape(-1,1)
    Y_train = test['total_tires'].values.reshape(-1,1)
    
    X_test = np.array(['2022-09-20', '2022-09-21', '2022-09-22', '2022-09-23', '2022-09-24', '2022-09-25', '2022-09-26']).reshape(-1,1).astype('datetime64[ns]')
    
    k = round(math.sqrt(len(X_train))/2)
    KNN_model = KNeighborsRegressor(n_neighbors=k).fit(X_train,Y_train)
    KNN_predict = KNN_model.predict(X_test) 
    KNN_predict = KNN_predict.reshape(1,-1)[0]
    
    predict_counter = 0
    for x in X_test:
        new_row = {'date': x, 'dc_name':'SAN JOSE', 'size_code':i, 'total_tires':KNN_predict[predict_counter]}
        final_results_san_jose = final_results_san_jose.append(new_row, ignore_index = True)
        predict_counter += 1
    
```

```python
## Sacramento ##
all_size_codes = sacramento['size_code'].unique()
final_results_sacramento = pd.DataFrame(columns = ['date', 'dc_name', 'size_code', 'total_tires'])

for i in all_size_codes:
    test = sacramento[sacramento['size_code'] == i] 
    X_train = test['date'].values.reshape(-1,1)
    Y_train = test['total_tires'].values.reshape(-1,1)
    
    X_test = np.array(['2022-09-20', '2022-09-21', '2022-09-22', '2022-09-23', '2022-09-24', '2022-09-25', '2022-09-26']).reshape(-1,1).astype('datetime64[ns]')
    
    k = round(math.sqrt(len(X_train))/2)
    KNN_model = KNeighborsRegressor(n_neighbors=k).fit(X_train,Y_train)
    KNN_predict = KNN_model.predict(X_test) 
    KNN_predict = KNN_predict.reshape(1,-1)[0]
    
    predict_counter = 0
    for x in X_test:
        new_row = {'date': x, 'dc_name':'SACRAMENTO', 'size_code':i, 'total_tires':KNN_predict[predict_counter]}
        final_results_sacramento = final_results_sacramento.append(new_row, ignore_index = True)
        predict_counter += 1
    
```

```python
## Oakland ##
all_size_codes = oakland['size_code'].unique()
final_results_oakland = pd.DataFrame(columns = ['date', 'dc_name', 'size_code', 'total_tires'])

for i in all_size_codes:
    test = oakland[oakland['size_code'] == i] 
    X_train = test['date'].values.reshape(-1,1)
    Y_train = test['total_tires'].values.reshape(-1,1)
    
    X_test = np.array(['2022-09-20', '2022-09-21', '2022-09-22', '2022-09-23', '2022-09-24', '2022-09-25', '2022-09-26']).reshape(-1,1).astype('datetime64[ns]')
    
    k = round(math.sqrt(len(X_train))/2)
    KNN_model = KNeighborsRegressor(n_neighbors=k).fit(X_train,Y_train)
    KNN_predict = KNN_model.predict(X_test) 
    KNN_predict = KNN_predict.reshape(1,-1)[0]
    
    predict_counter = 0
    for x in X_test:
        new_row = {'date': x, 'dc_name':'OAKLAND', 'size_code':i, 'total_tires':KNN_predict[predict_counter]}
        final_results_oakland = final_results_oakland.append(new_row, ignore_index = True)
        predict_counter += 1
    
```

```python
## Bakersfield ##
all_size_codes = bakersfield['size_code'].unique()
final_results_bakersfield = pd.DataFrame(columns = ['date', 'dc_name', 'size_code', 'total_tires'])


for i in all_size_codes:
    test = bakersfield[bakersfield['size_code'] == i] 
    X_train = test['date'].values.reshape(-1,1)
    Y_train = test['total_tires'].values.reshape(-1,1)
    
    X_test = np.array(['2022-09-20', '2022-09-21', '2022-09-22', '2022-09-23', '2022-09-24', '2022-09-25', '2022-09-26']).reshape(-1,1).astype('datetime64[ns]')
    
    k = round(math.sqrt(len(X_train))/2)
    KNN_model = KNeighborsRegressor(n_neighbors=k).fit(X_train,Y_train)
    KNN_predict = KNN_model.predict(X_test) 
    KNN_predict = KNN_predict.reshape(1,-1)[0]
    
    predict_counter = 0
    for x in X_test:
        new_row = {'date': x, 'dc_name':'BAKERSFIELD', 'size_code':i, 'total_tires':KNN_predict[predict_counter]}
        final_results_bakersfield = final_results_bakersfield.append(new_row, ignore_index = True)
        predict_counter += 1
    
```

```python
# putting all the results together into one dataframe
final1 = pd.concat([final_results_oakland,final_results_sacramento], axis = 0)
final2 = pd.concat([final1, final_results_san_jose], axis = 0)
final3 = pd.concat([final2, final_results_bakersfield], axis = 0)
```
